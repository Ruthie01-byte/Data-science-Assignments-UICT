```python
from google.colab import files
import pandas as pd

# Use the already uploaded file
carsales = pd.read_csv('Car_Sales (2).csv')
display(carsales.head())
```

```python
#displays the first 5 rows of the data set
carsales.head(5)
```

```python
carsales.info()
```

```python
# Check for null values
null_counts = carsales.isnull().sum()
display(null_counts)

# Calculate the percentage of null values
null_percentage = (null_counts / carsales.shape[0]) * 100
display(null_percentage)
```

```python
carsales.dropna(inplace=True)
carsales.isnull().sum()
```

```python
count_make = carsales['Make'].value_counts()
print(count_make)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
```

```python
#basic bar graph
plt.figure(figsize=(4,3))
count_make.plot(kind='bar', color='maroon')
plt.title('Number of cars by make', fontsize=15)
plt.xlabel('Make')
plt.xticks(rotation=60)
plt.ylabel('Number of cars')
plt.show()
```
